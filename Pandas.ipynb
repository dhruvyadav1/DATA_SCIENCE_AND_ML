# Sorting Series or a DataFrame

```python
movies = pd.read_csv('http://bit.ly/imdbratings')
```

```python
movies.head()
```

```python
#let's sort the movies by title name

movies['title'].sort_values().head()

#these are the first five 
```

```python
#to sort in descending order

movies['title'].sort_values(ascending=False).head()
```

```python
#let's sort the data based of the ratings

movies.sort_values(by='star_rating').head()
```

```python
#if there is a clash by a criteria we can have multiple criteria

movies.sort_values(['star_rating', 'duration','content_rating'],ascending=[True,True,True])
```

```python
#we can see for rating = 7.4 criteria is duration
```

# Indexing based of values

```python
#let's say we want to find movies with duration >=200

movies[movies['duration']>=200].head()
```

```python
#let's check the genre of movies with length >200

movies[movies['duration']>=200]['genre']
```

```python
# let's do the same with multiple conditions

#condition 1:  duration>=200

#condition 2:  genre = Drama
```

```python
movies[(movies['duration']>=20) & (movies['genre']=='Drama')].head()
```

```python
#IMPORTANT NOTE:

#All the conditions should be put in seperate '()'

# and can not be used we must use &
```

```python
#Working with Categorical Dtypes

df = pd.DataFrame({'ID':[100, 101, 102, 103], 'quality':['good', 'very good', 'good', 'excellent']})
```

```python
df
```

```python
from pandas.api.types import CategoricalDtype

cat_type = CategoricalDtype(categories=['good','very good','excellent'],ordered=True)
```

```python
#Now we can define the cat_type as a DataType
```

```python
df['quality'] = df['quality'].astype(cat_type)
```

```python
df.sort_values('quality')
```
